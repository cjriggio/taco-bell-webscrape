```python
import pandas as pd
import numpy as np 
from geopy.geocoders import Nominatim
```

```python
tacos = pd.read_csv('taco-bell.csv')
tacos.head()
```

```python
tacos.drop(['Unnamed: 0'], axis=1, inplace=True)
tacos.head()
```

```python
tacos.Address.iloc[0] + tacos.City.iloc[0]
```

```python
geolocator = Nominatim(user_agent="app_name_")
location = geolocator.geocode(tacos.Address.iloc[17], tacos.City.iloc[17])
location.latitude, location.longitude
```

```python
taco_dict = tacos.to_dict()
```

```python
taco_bell = 'Taco Bell'
```

```python
taco_dict['Address'][0] + taco_dict['City'][0] + ' ' + taco_dict['State'][0]
```

```python
lat = []
long = []
for i in range(0, len(tacos)):
    
    
    location = geolocator.geocode(taco_dict['Address'][i] + 
                                  taco_dict['City'][i] + 
                                  ' ' +
                                  taco_dict['State'][i]) 
                                
    if location is not None:
        lat.append(location.latitude)
        long.append(location.longitude)
    else:
        lat.append(0)
        long.append(0)
        print(i)

```

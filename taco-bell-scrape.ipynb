```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
def TacoBellScrape(city, state_abbreviation):
    city_ = city.lower()
    state = state_abbreviation.lower()
    url = 'https://locations.tacobell.com/{0}/{1}.html'.format(state, city_)
    
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    address = []
    city = []
    state = []
    zipcode = []
    
    for i in soup.find_all('address'):
        taco1 = i.find('span', attrs={'class':'c-address-street'})
        address.append(taco1.text)
    
    for i in soup.find_all('address'):
        taco2 = i.find('span', attrs={'class': 'c-address-city'})
        city.append(taco2.text)
    
    for i in soup.find_all('address'):
        taco3 = i.find('abbr', attrs={'class': 'c-address-state'}) 
        state.append(taco3.text)
    
    for i in soup.find_all('address'):
        taco4 = i.find('span', attrs={'class':'c-address-postal-code'})
        zipcode.append(taco4.text)

    df = pd.DataFrame([address, city, state, zipcode]).transpose()
    df.rename(columns={0:'Address', 
                       1:'City', 
                       2:'State', 
                       3:'Zipcode'},inplace=True)
    
    df['City'] = df['City'].map(lambda x: str(x)[:-1]) #removes comma from city column 
    
    return df
```

```python
TacoBellScrape('Brooklyn', 'NY')
```

```python
TacoBellScrape('Chicago', 'IL')
```
